```python
import pandas as pd
from pathlib import Path
import os
import json
```

```python
results_root_dir = '/home/gpubox3/projects/hf-exp/rafflesia/results'

results_file = 'convnextv2_base.fcmae_ft_in22k_in1k-224.csv'
```

```python
with open('/home/gpubox3/projects/hf-exp/model/classes-rafflesia-dataset-23986-2023-12-03.json', 'r') as f:
    class_to_index = json.load(f)

index_to_class = {int(index): cls_name for cls_name, index in class_to_index.items()}
```

```python
results_df = pd.read_csv(os.path.join(results_root_dir, results_file))
```

```python
results_df
```

```python
results_df['pred_class'] =results_df['index'].map(index_to_class)
```

```python
results_df['target_class'] = results_df['filename'].map(lambda x: x.split('/')[0])
```

```python
results_df.query("pred_class==target_class").shape[0] / results_df.shape[0]
```

```python
results_df.to_csv('./results/new_model_result_predictions.csv', index=False)
```
